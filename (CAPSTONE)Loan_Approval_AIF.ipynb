```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_column = ['Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
le = LabelEncoder()
for i in categorical_column:
    loan_train[i] = le.fit_transform(loan_train[i])
loan_train.head()
```

```python
x_train = loan_train.drop(["Loan_Status"], axis= 1)
xtrain = loan_train.drop(['Loan_ID'], axis= 1)
ytrain = loan_train["Loan_Status"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(xtrain,ytrain,test_size=0.2)
```

```python
reg = LogisticRegression()
reg.fit(x_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
a = round(reg.score(x_test,y_test)*100, 2)
print(f"The score of the model is {a}%")
```
